```python
#Now we utilize the elbow method to visualize the intertia for different values of K:
#Example
data = list(zip(x,y))
inertias = []
from sklearn.cluster import KMeans
for i in range(1,11):
    ke= KMeans(n_clusters = i)
    ke.fit(data)
    inertias.append(ke.inertia_)
plt.plot(range(1,11), inertias, marker = 'o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Import our kmeans algorithm
from sklearn.cluster import KMeans
```

```python
# Convert our dataframe to a form which is usable by kmeans
x = iris_df.iloc[:,[0,1]].values
# WCSS - Plot the elbow plot
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    # train the algorithm with the input data (x)
    kmeans.fit(x)
    # WCSS is also called inertia
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow curve')
plt.xlabel('number of k cluster')
plt.ylabel('wcss')
plt.show()
```
